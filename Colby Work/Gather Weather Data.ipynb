```python
dates = [str(x).split()[0] for x in pd.date_range(start='2010-01-01', end='2018-11-30', freq='D')]
years = [int(x.split('-')[0]) for x in dates]
months = [int(x.split('-')[1]) for x in dates]
new_df = pd.DataFrame({'year': years, 'month': months, 'date': dates})
new_df['date'] = pd.to_datetime(new_df['date'])
final = new_df.groupby(['year', 'month']).agg({'date': ['min', 'max']})['date'].reset_index()
final
```

```python
dates = []
mintemps = []
maxtemps = []
for i,row in final.iterrows():
    start_date = row["min"].date()
    end_date = row["max"].date()
    query_url = url + str(start_date) + "&enddate=" + str(end_date) + "&tp=24"
    response = requests.get(query_url).json()
    for x in range(len(response["data"]["weather"])):
        dates.append(response["data"]["weather"][x]["date"])
        mintemps.append(response["data"]["weather"][x]["mintempF"])
        maxtemps.append(response["data"]["weather"][x]["maxtempF"])
```

```python
d = {'Dates':dates,'MinTemp':mintemps,'MaxTemp':maxtemps}
weather = pd.DataFrame(d)
```

```python
weather.to_csv("weatherdata.csv")
```
